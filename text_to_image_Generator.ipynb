```python
from diffusers import StableDiffusionPipeline
import torch
HUGGING_FACE_TOKEN = "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"
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=HUGGING_FACE_TOKEN, torch_dtype=torch.float16)
pipe.to("cuda")
prompt = input("Enter Prompt of Image :")
print(f"Generating image for prompt: '{prompt}'")
image = pipe(prompt).images[0]
output_path = "generated_image.png"
image.save(output_path)
print(f"Image saved as {output_path}")
image.show()
```
